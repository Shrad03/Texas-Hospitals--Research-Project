```python
import pandas as pd
```

```python
df = pd.read_excel('/content/drive/MyDrive/ResearchProject.xlsx')
df.head()
```

```python
# Creating a condition to filter for the state of Texas
tx_condition = df['State'] == 'TX'

# Applying the condition to filter the DataFrame
tx_data = df[tx_condition]

# Displaying the filtered data
tx_data
```

```python
# Columns that are to be converted from object to float because the values in the columns are in percentages
columns_to_conv_obj = ['Relative\nprice for\noutpatient\nfacility\nservices', 'Relative\nprice for\ninpatient\nfacility\nservices']
#print
tx_data[columns_to_conv_obj]

```

```python
tx_data.loc[[2708, 2727, 2743, 3094, 3102]]['Simulated\nMedicare\nallowed\namount\nfor\noutpatient\nservices\n($\nmillions)']
```

```python
# # Custom function taken from https://stackoverflow.com/questions/12432663/what-is-a-clean-way-to-convert-a-string-percent-to-a-float
# def percent_to_float(x):
#     return float(x.strip('%'))/100

# # Pass to `converters` param as a dict...
# df = pd.read_csv(io.StringIO(temp1), sep='\s+',index_col=[0], converters={'col':p2f})
# df

tx_data[columns_to_conv_obj[0]] = tx_data[columns_to_conv_obj[0]].str.rstrip('%').astype('float')
tx_data[columns_to_conv_obj[1]] = tx_data[columns_to_conv_obj[1]].str.rstrip('%').astype('float')
tx_data[columns_to_conv_obj]
```

```python
tx_data['Relative\nprice for\noutpatient\nfacility\nservices'].plot(kind='hist') #checking the distribution since we only require outpatient information but not the inpatient
```

```python
#Null value imputation in the data

# Loop through all columns
for col in tx_data.columns:
    # Check if there are missing values
    if tx_data[col].isnull().any():
        # Check the data type of the column
        if tx_data[col].dtype == 'object':
            # Fill missing values in object columns with mode
            tx_data[col].fillna(tx_data[col].mode().iloc[0], inplace=True)
        elif pd.api.types.is_numeric_dtype(tx_data[col]):
            # Fill missing values in numeric columns with mean
            tx_data[col].fillna(tx_data[col].mean(), inplace=True)
```

```python
# Calculate Reimbursement Ratios
tx_data['Reimbursement Ratio'] = tx_data['Total\nprivate\nallowed\namount\nfor\noutpatient\nservices\n($\nmillions)'] / tx_data['Simulated\nMedicare\nallowed\namount\nfor\noutpatient\nservices\n($\nmillions)']
```

```python
# Select only float columns and round off to 2 digits
tx_data_float = tx_data.select_dtypes(include='float64').round(2)

# Combine the rounded float columns with non-float columns
sorted_hospitals = tx_data.drop(list(tx_data.select_dtypes(include='float64').columns), axis=1)
sorted_hospitals[list(tx_data.select_dtypes(include='float64').columns)] = tx_data_float

# Sort the DataFrame based on 'Reimbursement Ratio' in descending order
sorted_hospitals = sorted_hospitals.sort_values(by='Reimbursement Ratio', ascending=False)
sorted_hospitals[['Hospital name', 'City', 'Reimbursement Ratio']].head(10)
```
